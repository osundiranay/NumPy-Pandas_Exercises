# 오류 상황 2
    fatal: The current branch main has no upstream branch.
    To push the current branch and set the remote as upstream, use

    git push --set-upstream origin main
git push 명령어를 입력하였을 때 나타나는 경우가 있다.<br>
위와 같은 에러가 나는 이유는 정확하게 원격저장소 이름을 언급해주지 않았다는 이야기이다.<br>
따라서 git remote -v 를 통해 현재 저장된 원격 저장소 이름을 찾은 뒤 원격 저장소를 명시하여 적어주면 해결된다.<br>
> git remote -v

> git push origin master (origin: 원격저장소 이름)

```python
!git remote -v
```

```python
!git push origin master
```
