## Solution

# 3\. How to convert the index of a series into a column of a dataframe?
Difficulty Level: L1<br>

Convert the series ser into a dataframe with its index as another column on the dataframe.<br>

> input:<br>

    mylist = list('abcedfghijklmnopqrstuvwxyz')
    myarr = np.arange(26)
    mydict = dict(zip(mylist, myarr))
    ser = pd.Series(mydict)

## Solution

> My solution

```python
mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser = pd.Series(mydict)

pd.DataFrame(ser).reset_index()
```

> Solution

```python
# Input
mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser = pd.Series(mydict)

# Solution
df = ser.to_frame().reset_index()
print(df.head())
```

> \# Study Point
- `(Series).to_frame()`은 Series 데이터를 DataFrame 형식의 데이터로 변환해준다.
- `.reset_index()`는 index를 기본값으로 초기화 해준다.
    - 이때 기존에 있던 index는 첫 번째 column이 된다.
    - 괄호 안에 `drop=True`라는 옵션을 주면 기존의 index를 버리고 재배열 해준다.

```python
# Example
srr = pd.Series({'Name':'Mike', 'Age':20, 'Sex':'Male'})

df = srr.to_frame()
df
```

```python
# Example 2
df.reset_index()
```

```python
df.reset_index(drop=True)
```
